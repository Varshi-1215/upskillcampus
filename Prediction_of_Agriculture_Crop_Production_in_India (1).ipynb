```python
df = df.dropna()
```

```python
df.isnull().sum()
```

```python
years = df.columns[1:]
df[years] = df[years].apply(pd.to_numeric)
```

```python
df.dtypes
```

```python
import matplotlib.pyplot as plt

crop = df.iloc[0, 0]
values = df.iloc[0, 1:]

plt.plot(years, values)
plt.title(crop + " Production Trend")
plt.xlabel("Years")
plt.ylabel("Production")
plt.xticks(rotation=45)
plt.show()
```

```python
top5 = df.set_index('Crop').mean(axis=1).sort_values(ascending=False).head(5)
top5
```

```python
import matplotlib.pyplot as plt

top5.plot(kind='bar')
plt.title("Top 5 Crops by Average Production")
plt.xlabel("Crop")
plt.ylabel("Average Production")
plt.xticks(rotation=45)
plt.show()
```
